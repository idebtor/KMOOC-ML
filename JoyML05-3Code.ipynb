## 2. 클래스: 설계

```python
class Car:
    def __init__(self, length, weight):
        self.length = length
        self.weight = weight
        self.speed = 0
    def forward(self):
        print("Move Forward!")
    def backward(self):
        print("Move Backward!")
    def move(self, owner):
        owner.pos += 10
    def speedUP(self):
        self.speed += 10
    def change(self, length, weight):
        self.length = length
        self.weight = weight
```

```python
myCar = Car(4, 1380)
myCar.change(3.5, 1200)
print(myCar.length)
print(myCar.weight)
```

```python
class Person:
    def __init__(self):
        self.pos = 0
    def start(self, Car):
        print("Whoom!!")  
```
