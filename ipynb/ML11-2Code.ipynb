## 5. 원-핫-인코딩: one-hot encoding

```python
def one_hot_encoding(y, n_y, modified = True):
    yhot = np.eye(n_y)[np.array(y, dtype = 'int32').flatten()]
    if modified:
        yhot[yhot == 0] = 0.01
        yhot[yhot == 1] = 0.99
    return yhot
```
