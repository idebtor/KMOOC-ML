```python
import matplotlib.pyplot as plt
%matplotlib inline

def mileToKm(x):
    """x 마일을 Km로 변환하여 반환"""
    return 1.61 * x

def plotMileToKm(x,y):
    """x, y의 값들을 그래프로 출력"""
    plt.figure()
    plt.plot(x,y)
    plt.title('Mile to Km')
    plt.xlabel('Mile')
    plt.ylabel('Km')
    plt.show()

for mile in range(0, 5):
        print('{}mile:{}km'.format(mile, mileToKm(mile)))
```
